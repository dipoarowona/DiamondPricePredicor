```python
#we will be using machine learning 
#using a regression model 
#machine learning - more than 10k rows to 100k rows
#deep learning is 100K or more

import pandas as pd

df = pd.read_csv("diamonds.csv", index_col = 0)
df.head()
```

```python
#in machine learning everything must be in terms of numbers so cut color and clarity have to find a way to be converted into a numerical value
#to set a column of data that are stings to arbitrary numerical values the following can be used (the numbers dont mean anything)
#df["ROWNAME"].astype("category").cat.codes
#but for this you want to keep the order of premium good, etc
 

cut_class_dict= {"Fair":1, "Good":2, "Very Good":3, "Premium":4, "Ideal":5}
clarity_dict = {"I3": 1, "I2": 2, "I1": 3, "SI2": 4, "SI1": 5, "VS2": 6, "VS1": 7, "VVS2": 8, "VVS1": 9, "IF": 10, "FL": 11}
color_dict = {"J": 1,"I": 2,"H": 3,"G": 4,"F": 5,"E": 6,"D": 7}

#map the dictionaries met to the corresponding column 

df["cut"] = df["cut"].map(cut_class_dict)
df["color"] = df["color"].map(color_dict)
df["clarity"] = df["clarity"].map(clarity_dict)

df.head()
```

```python
import sklearn
from sklearn import svm, preprocessing
#this is imports necessary so that the machine learning can be done
#shuffle the data because the order might induce a bias

df = sklearn.utils.shuffle(df)#shuffles dataset

#in machine learning x is defined  as the feature set that point to a label
# and y is known as the label
# in this case the feature set is everything but price and price is the label

X = df.drop("price", axis=1).values#we want it converted to values
 #scaling down the data allows for easier processing and just overall helps simplify the data
X = preprocessing.scale(X)
y = df["price"].values

#now we want to section the data into test and train subsets
# you dont want to test your algorithing with the same data it is training with that is why we create a section the algo will not see until tested

test_size = 200 

X_train = X[:-test_size]
y_train = y[:-test_size]
#training on everything but the last 200 

X_test = X[-test_size:]
y_test = y[-test_size:]

#determine the classifier
clf = svm.SVR(kernel="linear") #changes depening on type of regression
clf.fit(X_train, y_train) #trains machine



```

```python
#to test the model we use score method
clf.score(X_test, y_test)
```

```python
#display predictions

for X,y in zip(X_test, y_test):
    print(f'Model: {clf.predict([X])[0]},\tActual: {y}')
    #For predict you must always pass a list and index the list, 0 if you are only expecting one result
```

```python
#retesting with another kernal/linear regression model 

clf = svm.SVR(kernel = "rbf")
clf.fit(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

```python
for X,y in zip(X_test, y_test):
    print(f"Model: {clf.predict([X])[0]},\tActual:{y}")
```
